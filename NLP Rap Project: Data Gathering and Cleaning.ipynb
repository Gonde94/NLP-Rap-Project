```python
from bs4 import BeautifulSoup
import requests
import pickle
```

```python
#add artists to UrlList

urlList = ["https://genius.com/artists/Wu-tang-clan",
           "https://genius.com/artists/A-tribe-called-quest",
           "https://genius.com/artists/Nwa",
           "https://genius.com/artists/Outkast",
           "https://genius.com/artists/Nas",
           "https://genius.com/artists/The-notorious-big",
           "https://genius.com/artists/2pac",
           "https://genius.com/artists/Logic",
           "https://genius.com/artists/Jay-z-and-kanye-west",
           "https://genius.com/artists/Tyler-the-creator",
           "https://genius.com/artists/Immortal-technique",
           "https://genius.com/artists/Kendrick-lamar"]

artistnames = ['Wu-Tang Clan','A Tribe Called Quest','N.W.A.','Outkast','Nas','The Notorious B.I.G.',
               '2pac','Logic','Jay-z and Kanye West','Tyler the Creator','Immortal Technique', 'Kendrick Lamar']
```

```python
#Grab top 10 songs

songList = []

for e in urlList:
    page = requests.get(e)
    soup = BeautifulSoup(page.content, "html.parser")
    
    songLink = soup.find("div", class_="mini_card_grid")
    
    for link in songLink.find_all('a'):
        URL = link.get('href')
        print(URL)
        songList.append(URL)
```

```python
#Remove songs where artist is not primary artist
unwanted = [33,45,49,59,75,84,92,107,111,117,119]

for e in sorted(unwanted,reverse=True):
    del songList[e]
```

```python
#Grab title of each song

titles = []

for e in songList:
    page = requests.get(e)
    soup = BeautifulSoup(page.content, "html.parser")
    
    titles.append(soup.find("h1").text)
```

```python
#Album/Track release year

albumyear = []

for e in songList:
    page = requests.get(e)
    soup = BeautifulSoup(page.content, "html.parser")
    
    year = soup.find('div', class_="SongHeaderVariantdesktop__Column-sc-12tszai-1 SongHeaderVariantdesktop__Right-sc-12tszai-4 jryaLr").get_text()
    
    yearint = re.findall('\d+', year)
    
    albumyear.append(yearint[1])
```

```python
#Grab lyrics for all songs

lyricslist = []

for e in songList:
    page = requests.get(e)
    soup = BeautifulSoup(page.content, "html.parser")
    
    lyricspersong = soup.find("div", id='lyrics-root').get_text(', ')
    
    lyricslist.append(lyricspersong)
```

```python
artistdictionary = {}
```

```python
for e in artistnames:
    artistdictionary[e] = None
```

```python
artistdictionary
```

```python
#combine lyrics for each artist
artistdictionary.update({'Wu-Tang Clan': lyricslist[:9]})
artistdictionary.update({'A Tribe Called Quest': lyricslist[10:19]})
artistdictionary.update({'N.W.A.': lyricslist[20:29]})
artistdictionary.update({'Outkast': lyricslist[30:39]})
artistdictionary.update({'Nas': lyricslist[40:49]})
artistdictionary.update({'The Notorious B.I.G.': lyricslist[50:59]})
artistdictionary.update({'2pac': lyricslist[60:69]})
artistdictionary.update({'Logic': lyricslist[70:79]})
artistdictionary.update({'Jay-z and Kanye West': lyricslist[80:89]})
artistdictionary.update({'Tyler the Creator': lyricslist[90:99]})
artistdictionary.update({'Immortal Technique': lyricslist[100:109]})
artistdictionary.update({'Kendrick Lamar': lyricslist[110:119]})
```

```python
#combine all the text into one, rather than separate elements 
def combine_text(list_of_text):
    combined_text = ' '.join(list_of_text)
    return combined_text
```

```python
data_combined = {key: [combine_text(value)] for (key, value) in artistdictionary.items()}
```

```python
#finally, put this all in its own dataframe
import pandas as pd

data_df = pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns = ['Lyrics']
data_df = data_df.sort_index()
```

# DATA CLEANING

```python
#Clean up round 1
import string

def clean_up_1(text):
    text = text.lower()
#    text = re.sub('\[.*-?\]','',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    text = re.sub('[''""...-]','',text)
    text = re.sub('\n','',text)
    return text
    
round_1 = lambda x: clean_up_1(x)
```

```python
data_clean = pd.DataFrame(data_df.Lyrics.apply(round_1))
```

```python
pd.set_option('max_colwidth',150)
data_clean
```

```python
 def clean_up_2(text):  
    text = text.replace('lyrics', '')
    text = text.replace('produced by', '')
    text = text.replace('intro', '')
    text = text.replace('verse', '')
    text = text.replace('chorus', '')
    text = text.replace('embed', '')
    return text

round_2 = lambda x: clean_up_2(x)
```

```python
data_clean = pd.DataFrame(data_clean.Lyrics.apply(round_2))
data_clean
```

# COUNT VECTORIZER

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(data_clean['Lyrics'])
data_dtm = pd.DataFrame(data_cv.toarray(),columns=cv.get_feature_names_out())
data_dtm.index = data_clean.index
```
